```python
import gzip
import pandas as pd
import string
import os
```

```python
print(os.getcwd())
filepath= os.getcwd()# path for unzip
```

```python
# Getting all files to unzip
file_names = list(names.name for names \
        in os.scandir(filepath+'/zippedData/'))
print(file_names)
df_names = []
for name in file_names:
    df_names.append(name.replace(".csv.gz","").replace(".tsv.gz", "").replace(".","_"))
print(df_names) # Creates list for names of future dataframes in string format
```

```python
# Opens zip files and turns each into database
for index, name in enumerate(file_names):
    # Opens the file
    obj = gzip.open(f"{filepath}/zippedData/{name}")
    
    # Tries to read it as csv and if not then tells us to manually open it.
    try:
        obj_df = pd.read_csv(obj)
        exec(f'{df_names[index]}_df = obj_df')
    except:
        print('DO THIS ONE BY HAND {}'.format(name))
       
    # Shows name of databases
    print(df_names[index])
```

### In the loop above, we open every zip file and create a dataframe for it. the name of the dataframe is df_names+_df. all files that couldn't have this done to them, we got a print statement telling us to do it by hand

```python
# Sums the orders for the movie worldwide
title_akas_ordering = imdb_title_akas_df.groupby(imdb_title_akas_df['title_id']).sum()
```

```python
# Converts region column to list of regions
title_akas_regions = imdb_title_akas_df[['title_id','ordering','title', 'region']]\
                .dropna().groupby(imdb_title_akas_df['title_id'])['region'].apply(','.join)
```

```python
# Converts language column to list of languages (NOT UNIQUE)
title_akas_langs = imdb_title_akas_df[['title_id','ordering','title', 'language']]\
                .dropna().groupby(imdb_title_akas_df['title_id'])['language'].apply(','.join)
```

```python
# Creates base df for merges
title_akas_originals = imdb_title_akas_df[['title_id','title']][(imdb_title_akas_df['is_original_title']>0)]
```

```python
# Merges last four cells to one df
worldwide_title_akas_imdb = pd.merge(left=title_akas_ordering, right=title_akas_regions, on='title_id')\
                            .join(title_akas_langs, on='title_id')
worldwide_region_df = pd.merge(left=worldwide_title_akas_imdb, right=title_akas_originals, on='title_id')
```

```python
# Function to convert money from string to int
def dfcolumn_convert_money_to_int(column):
    """
    Takes in a dataframe.column of object value with the format meant to 
    represent monetary value and returns the new column values of int type
    """
    try:
        new_col = column.apply(lambda x: int(x.replace("$","").replace(",","")))
        return new_col
    except:
        print('Elements in this series are currently of {} type'\
              .format(type(column[0])))
        return
```

```python
tn_movie_budgets_df['production_budget'] = dfcolumn_convert_money_to_int(tn_movie_budgets_df['production_budget'])
```
